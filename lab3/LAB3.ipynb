```python
#TASK 1 Convert the following searching algorithms into agent-based models: ● Depth-Limited Search (DLS): Implement as a Goal-Based Agent to explore the graph up to a specified depth limit.
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': [],
    'F': ['I'],
    'G': [],
    'H': [],
    'I': []
}

class Environment:
    def __init__(self, graph):
        self.graph = graph

    def get_percept(self, node):
        return node


class GoalBasedAgent:
    def __init__(self, goal, start):
        self.goal = goal
        self.start = start

    def formulate_goal(self, percept):
        if percept == self.goal:
            return "Goal reached"
        else:
            return "Searching"

    def dls(self, graph, start, goal, depth_limit):
        visited = []

        def dfs(node, depth):
            if depth > depth_limit:
                return None
            visited.append(node)
            if node == goal:
                print(f"Goal found with DLS. Path: {visited}")
                return visited
            for neighbor in graph.get(node, []):
                if neighbor not in visited:
                    path = dfs(neighbor, depth + 1)
                    if path:
                        return path
            visited.pop()
            return None

        result = dfs(start, 0)
        if result:
            return f"Goal {goal} found within depth limit {depth_limit}!"
        else:
            return f"Goal {goal} not found within depth limit {depth_limit}."

    def act(self, percept, graph, depth_limit):
        goal_status = self.formulate_goal(percept)
        if goal_status == "Goal reached":
            return f"Goal {self.goal} found!"
        else:
            return self.dls(graph, self.start, self.goal, depth_limit)


def run_agent(agent, environment, start_node, depth_limit):
    percept = environment.get_percept(start_node)
    action = agent.act(percept, environment.graph, depth_limit)
    print(action)


start_node = 'A'
goal_node = 'I'

depth_limit = 3

agent = GoalBasedAgent(goal_node, start_node)
environment = Environment(tree)

run_agent(agent, environment, start_node, depth_limit)
```

```python
# TASK 2 Traveling Salesman Problem:
graph = {
    '1': {'2': 10, '3': 15, '4': 20},
    '2': {'1': 10, '3': 35, '4': 25},
    '3': {'1': 15, '2': 35, '4': 30},
    '4': {'1': 20, '2': 25, '3': 30}
}

class TSP_UCS:
    def __init__(self, graph, start):
        self.graph = graph
        self.start = start
        self.num_cities = len(graph)
        self.min_cost = float('inf')
        self.best_path = []

    def ucs(self):
        """Finds the shortest Hamiltonian cycle using UCS without external libraries."""
        frontier = [(0, [self.start])]  # (cost, path)

        while frontier:
            # Manually sort frontier based on cost (simulating priority queue)
            frontier.sort(key=lambda x: x[0])
            current_cost, path = frontier.pop(0)
            current_city = path[-1]

            # If all cities are visited, check return cost to start
            if len(path) == self.num_cities:
                if self.start in self.graph[current_city]:  # Ensure cycle completion
                    cycle_cost = current_cost + self.graph[current_city][self.start]
                    if cycle_cost < self.min_cost:
                        self.min_cost = cycle_cost
                        self.best_path = path + [self.start]
                continue


            for neighbor, cost in self.graph[current_city].items():
                if neighbor not in path:
                    new_cost = current_cost + cost
                    new_path = path + [neighbor]
                    frontier.append((new_cost, new_path))
        if self.best_path:
           return self.best_path, self.min_cost
        else:
          "No solution found"


tsp_solver = TSP_UCS(graph, '1')
best_route, min_cost = tsp_solver.ucs()
print(f"Optimal TSP Route: {best_route}, Minimum Cost: {min_cost}")
```

```python
# TASK  3
tree = {
  'A': ['B', 'C'],
  'B': ['D', 'E'],
  'C': ['F', 'G'],
  'D': ['H'],
  'E': [],
  'F': ['I'],
  'G': [],
  'H': [],
  'I': []
}

graph = {
  'A': ['B', 'C'],
  'B': ['A', 'D', 'E'],
  'C': ['A', 'F', 'G'],
  'D': ['B', 'H'],
  'E': ['B'],
  'F': ['C', 'I'],
  'G': ['C'],
  'H': ['D'],
  'I': ['F']
}


def dlstree(node, goal, depth, path):
  if depth == 0:
    return False
  if node == goal:
    path.append(node)
    return True
  if node not in tree:
    return False
  for child in tree[node]:
    if dlstree(child, goal, depth - 1, path):
      path.append(node)
      return True
  return False

def iterative_deepening_tree(start, goal, max_depth):
  for depth in range(max_depth + 1):
    print(f"Depth: {depth}")
    path = []
    if dlstree(start, goal, depth, path):
      print("\nPath to goal:", " → ".join(reversed(path)))
      return
  print("Goal not found within depth limit.")


def dls_graph(node, goal, depth, path):
  if depth == 0:
    return False
  if node == goal:
    path.append(node)
    return True
  if node not in graph:
    return False
  for child in graph[node]:
    if dls_graph(child, goal, depth - 1, path):
      path.append(node)
      return True
  return False

def iterative_deepening_graph(start, goal, max_depth):
  for depth in range(max_depth + 1):
    print(f"Depth: {depth}")
    path = []
    if dls_graph(start, goal, depth, path):
      print("\nPath to goal:", " → ".join(reversed(path)))
      return
  print("Goal not found within depth limit.")


start_node = 'A'
goal_node = 'I'
max_search_depth = 5
print('CHECKING FOR TREE')
iterative_deepening_tree(start_node, goal_node, max_search_depth)
print('\nCHECKING FOR GRAPH')
iterative_deepening_graph(start_node, goal_node,max_search_depth)
```

```python
#TASK 4
grid = [
    ['S', 0, 1, 0, 0],
    [0, 1, 0, 1, 0],
    [0, 0, 0, 1, 'G'],
    [1, 1, 0, 0, 0]
]

def create_graph(grid):
    graph = {}
    rows, cols = len(grid), len(grid[0])
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for i in range(rows):
        for j in range(cols):
            if grid[i][j] != 1:
                neighbors = []
                for dx, dy in directions:
                    nx, ny = i + dx, j + dy
                    if 0 <= nx < rows and 0 <= ny < cols and grid[nx][ny] != 1:
                        neighbors.append((nx, ny))
                graph[(i, j)] = neighbors
    return graph

def bfs(graph, start, goal):
    queue = [(start, [start])]
    visited = set()

    while queue:
        current, path = queue.pop(0)

        if current == goal:
            return path

        if current not in visited:
            visited.add(current)
            for neighbor in graph.get(current, []):
                if neighbor not in visited:
                    queue.append((neighbor, path + [neighbor]))

    return "No path found"


def find_start_goal(grid):
  start, goal = None, None
  for i in range(len(grid)):
      for j in range(len(grid[0])):
          if grid[i][j] == 'S':
              start = (i, j)
          elif grid[i][j] == 'G':
              goal = (i, j)
  return start, goal


start, goal = find_start_goal(grid)
graph = create_graph(grid)
shortest_path = bfs(graph, start, goal)

print(f"Shortest Path from S to G: {shortest_path}")
```
